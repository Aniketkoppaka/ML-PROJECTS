## Hyperparameter tuning



```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint

param_dist_dt = {
    'max_depth': sp_randint(1, 20),
    'min_samples_split': sp_randint(2, 20),
    'min_samples_leaf': sp_randint(1, 20),
    'criterion': ['gini', 'entropy']
}

param_dist_rf = {
    'n_estimators': sp_randint(10, 200),
    'max_depth': sp_randint(1, 20),
    'min_samples_split': sp_randint(2, 20),
    'min_samples_leaf': sp_randint(1, 20),
    'bootstrap': [True, False],
    'criterion': ['gini', 'entropy']
}

random_search_dt = RandomizedSearchCV(models['Decision Tree'], param_distributions=param_dist_dt, n_iter=100, cv=kf, scoring='f1', random_state=42, n_jobs=-1)

random_search_dt.fit(X_train, y_train)

print("Best parameters for Decision Tree:", random_search_dt.best_params_)
print("Best F1-score for Decision Tree:", random_search_dt.best_score_)

random_search_rf = RandomizedSearchCV(models['Random Forest'], param_distributions=param_dist_rf, n_iter=100, cv=kf, scoring='f1', random_state=42, n_jobs=-1)

random_search_rf.fit(X_train, y_train)

print("Best parameters for Random Forest:", random_search_rf.best_params_)
print("Best F1-score for Random Forest:", random_search_rf.best_score_)
```

**Reasoning**:
Train the Decision Tree and Random Forest models with the best hyperparameters found and evaluate their performance on the test data using precision, recall, F1-score, and AUC-ROC to compare with the default models.

## Model interpretation



```python
models_evaluated = {
    'Logistic Regression (Default)': LogisticRegression(max_iter=1000).fit(X_train, y_train),
    'Support Vector Machine (Default)': SVC(probability=True).fit(X_train, y_train),
    'Decision Tree (Default)': DecisionTreeClassifier().fit(X_train, y_train),
    'Random Forest (Default)': RandomForestClassifier().fit(X_train, y_train),
    'Gradient Boosting (Default)': GradientBoostingClassifier().fit(X_train, y_train),
    'Decision Tree (Tuned)': best_dt_model,
    'Random Forest (Tuned)': best_rf_model
}

performance_metrics = {}

for model_name, model in models_evaluated.items():
    y_pred = model.predict(X_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    auc = 'N/A'
    if hasattr(model, 'predict_proba'):
        y_proba = model.predict_proba(X_test)[:, 1]
        auc = roc_auc_score(y_test, y_proba)

    performance_metrics[model_name] = {
        'Precision': precision,
        'Recall': recall,
        'F1-score': f1,
        'AUC-ROC': auc
    }

print("Model Performance on Test Set:")
for model_name, metrics in performance_metrics.items():
    print(f'{model_name}:')
    for metric, score in metrics.items():
        print(f'  {metric}: {score:.4f}')

best_model_name = None
best_f1 = -1
best_auc = -1

for model_name, metrics in performance_metrics.items():
    if metrics['F1-score'] != 'N/A' and metrics['F1-score'] > best_f1:
        best_f1 = metrics['F1-score']
        best_model_name = model_name
        if metrics['AUC-ROC'] != 'N/A':
             best_auc = metrics['AUC-ROC']
        else:
             best_auc = -1

    elif metrics['F1-score'] != 'N/A' and metrics['F1-score'] == best_f1:
         if metrics['AUC-ROC'] != 'N/A' and metrics['AUC-ROC'] > best_auc:
              best_auc = metrics['AUC-ROC']
              best_model_name = model_name


print(f"\nBest performing model based on F1-score and AUC-ROC on the test set: {best_model_name}")

if "Decision Tree" in best_model_name or "Random Forest" in best_model_name or "Gradient Boosting" in best_model_name:
    best_model = models_evaluated[best_model_name]
    if hasattr(best_model, 'feature_importances_'):
        importances = best_model.feature_importances_
        feature_names = X_test.columns
        feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
        feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)

        plt.figure(figsize=(12, 8))
        sns.barplot(x='importance', y='feature', data=feature_importance_df.head(15))
        plt.title(f'Top 15 Feature Importances for {best_model_name}')
        plt.show()

        print("\nTop 15 Most Important Features:")
        display(feature_importance_df.head(15))

elif "Logistic Regression" in best_model_name:
    best_model = models_evaluated[best_model_name]
    coefficients = best_model.coef_[0]
    feature_names = X_test.columns
    coefficient_df = pd.DataFrame({'feature': feature_names, 'coefficient': coefficients})
    coefficient_df['abs_coefficient'] = abs(coefficient_df['coefficient'])
    coefficient_df = coefficient_df.sort_values('abs_coefficient', ascending=False)

    plt.figure(figsize=(12, 8))
    sns.barplot(x='coefficient', y='feature', data=coefficient_df.head(15))
    plt.title(f'Top 15 Feature Coefficients for {best_model_name}')
    plt.show()

    print("\nTop 15 Features by Absolute Coefficient Value:")
    display(coefficient_df.head(15))
```

## Summary:

### Data Analysis Key Findings

*   Initial data exploration revealed that numerical features like 'Max\_heart\_rate' show a positive correlation with the target variable, while 'age', 'resting\_blood\_pressure', and 'cholestoral' show weaker negative correlations.
*   Box plots indicated differences in the distributions of 'Max\_heart\_rate' and 'oldpeak' between individuals with and without heart disease.
*   Engineered features, including the interaction between 'age' and 'Max\_heart\_rate' and the square of 'oldpeak', were created and added to the dataset.
*   Initial model training showed that Decision Tree and Random Forest models achieved the highest accuracy (approximately 98.54\%) with default hyperparameters.
*   Cross-validation results using F1-score indicated that Random Forest (default) had a strong performance with an F1-score of 0.9659, while Gradient Boosting also performed well with an F1-score of 0.9552.
*   Hyperparameter tuning using RandomizedSearchCV improved the cross-validated F1-score for both Decision Tree (0.9474) and Random Forest (0.9671).
*   On the test set, the Default Random Forest model achieved a perfect Precision of 1.0000, a high Recall of 0.9709, an F1-score of 0.9852, and an AUC-ROC of 1.0000. The Tuned Random Forest also performed very well with similar metrics.
*   Feature importance analysis of the Default Random Forest model identified 'chest\_pain\_type\_Typical angina', 'thalassemia\_Reversable Defect', 'Max\_heart\_rate', and 'vessels\_colored\_by\_flourosopy\_Zero' as the most important features for predicting heart disease. Engineered features like 'oldpeak\_sq' and 'age\_Max\_heart\_rate\_interaction' were also found to be important.
